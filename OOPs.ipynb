## **THEORY QUESTIONS**
## **Q1. What is Object-Oriented Programming (OOP)?**

Answer:
OOP is a programming style where we use classes and objects to write code.
It helps in organizing programs and makes them easier to reuse and maintain.
It works with concepts like inheritance, encapsulation, abstraction, and polymorphism.

## **Q2. What is a class in OOP?**

**Answer:**
A class is like a blueprint or template that defines how objects will look and behave.
It contains attributes (data) and methods (functions).



## **Q3. What is an object in OOP?**

**Answer:**
An object is an instance of a class.
It is created from a class and has its own data.



## **Q4. What is the difference between abstraction and encapsulation?**

**Answer:**

* Abstraction hides unnecessary details and shows only the important parts.
* Encapsulation protects data by keeping it private and using methods to access it.


## **Q5. What are dunder methods in Python?**

**Answer:**
Dunder methods are special methods in Python that begin and end with double underscores.
Examples: `__init__()`, `__str__()`, `__add__()`.
They allow us to customize how objects behave.



## **Q6. Explain the concept of inheritance in OOP.**

**Answer:**
Inheritance allows one class (child) to take attributes and methods from another class (parent).
It helps with code reuse.



## **Q7. What is polymorphism in OOP?**

**Answer:**
Polymorphism means the same method name can behave differently depending on the object calling it.


## **Q8. How is encapsulation achieved in Python?**

**Answer:**
Encapsulation is done using private variables (like `_balance` or `__balance`)
and providing getter/setter methods to control access.



## **Q9. What is a constructor in Python?**

**Answer:**
A constructor is the `__init__()` method.
It is called automatically when an object is created and initializes object attributes.



## **Q10. What are class and static methods in Python?**

**Answer:**

* Class method uses `@classmethod` and takes `cls` as the first argument.
* Static method uses `@staticmethod` and takes no class or object reference.



## **Q11. What is method overloading in Python?**

**Answer:**
Python does not support method overloading directly,
but we can mimic it using default parameters inside a method.



## **Q12. What is method overriding in OOP?**

**Answer:**
Method overriding happens when a child class provides a new implementation
for a method already defined in the parent class.


## **Q13. What is a property decorator in Python?**

**Answer:**
`@property` lets us access a method like an attribute.
It is used to control how values are read or modified.



## **Q14. Why is polymorphism important in OOP?**

**Answer:**
Polymorphism makes code flexible, reusable, and easier to maintain
because one method name can work differently for different objects.



## **Q15. What is an abstract class in Python?**

**Answer:**
An abstract class cannot be instantiated and contains abstract methods
that must be implemented by child classes.
We create it using `ABC` and `@abstractmethod`.



## **Q16. What are the advantages of OOP?**

**Answer:**

* Code reusability
* Better organization
* Easier to maintain
* More secure
* Represents real-world objects



## **Q17. What is the difference between a class variable and an instance variable?**

**Answer:**

* Class variable → shared by all objects
* Instance variable → unique for each object



## **Q18. What is multiple inheritance in Python?**

**Answer:**
Multiple inheritance means a class can inherit from more than one parent class.



## **Q19. Explain the purpose of `__str__()` and `__repr__()` in Python.**

**Answer:**

* `__str__()` → returns a user-friendly string
* `__repr__()` → returns a developer-friendly string



## **Q20. What is the significance of the `super()` function in Python?**

**Answer:**
`super()` is used in a child class to call methods from the parent class.



## **Q21. What is the significance of the `__del__` method in Python?**

**Answer:**
`__del__()` is a destructor that is automatically called when an object is deleted.



## **Q22. What is the difference between @staticmethod and @classmethod in Python?**

**Answer:**

* `@staticmethod`: no reference to class or object.
* `@classmethod`: takes `cls` as the first parameter.



## **Q23. How does polymorphism work in Python with inheritance?**

**Answer:**
Polymorphism works when a child class overrides a method of its parent class,
and each object calls its own version of the method.



## **Q24. What is method chaining in Python OOP?**

**Answer:**
Method chaining means calling multiple methods in a single line
because each method returns `self`.

Example:

```
obj.setName().setAge().setAddress()
```



## **Q25. What is the purpose of the `__call__` method in Python?**

**Answer:**
`__call__()` allows an object to be used like a function.
If a class defines it, then:

```
obj()
```

will automatically call `obj.__call__()`.

````python
#Q2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return 3.14 * self.r * self.r

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h


c = Circle(5)
r = Rectangle(4, 6)
print("Circle Area:", c.area())
print("Rectangle Area:", r.area())
````

````python
#Q3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.
class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, type, brand, battery):
        super().__init__(type, brand)
        self.battery = battery


e = ElectricCar("Car", "Tesla", "85 kWh")
print(e.type, e.brand, e.battery)
````

````python
#Q4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.
class Bird:
    def fly(self):
        print("Bird is flying")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high")

class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly")


for b in [Sparrow(), Penguin()]:
    b.fly()
````

````python
#Q5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance")

    def check_balance(self):
        return self.__balance


account = BankAccount(5000)
account.deposit(1000)
account.withdraw(3000)
print("Balance:", account.check_balance())
````

````python
#Q6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().
class Instrument:
    def play(self):
        print("Instrument playing")

class Guitar(Instrument):
    def play(self):
        print("Guitar is playing")

class Piano(Instrument):
    def play(self):
        print("Piano is playing")


for inst in [Guitar(), Piano()]:
    inst.play()
````

````python
#Q7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b


print(MathOperations.add_numbers(10, 5))
print(MathOperations.subtract_numbers(10, 5))
````

````python
#Q8. Implement a class Person with a class method to count the total number of persons created.
class Person:
    count = 0

    def __init__(self, name):
        self.name = name
        Person.count += 1

    @classmethod
    def total_persons(cls):
        return cls.count


p1 = Person("A")
p2 = Person("B")
print("Total Persons:", Person.total_persons())
````

````python
#Q9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the fraction as "numerator/denominator".
class Fraction:
    def __init__(self, n, d):
        self.n = n
        self.d = d

    def __str__(self):
        return f"{self.n}/{self.d}"

# Testing
f = Fraction(3, 4)
print(f)
````

````python
#Q10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Testing
v1 = Vector(2, 3)
v2 = Vector(4, 1)
print(v1 + v2)
````

````python
#Q12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

# Testing
s = Student("Rahul", [85, 90, 95])
print("Average Grade:", s.average_grade())
````

````python
#Q14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.
class Employee:
    def calculate_salary(self, hours, rate):
        return hours * rate

class Manager(Employee):
    def calculate_salary(self, hours, rate, bonus):
        return super().calculate_salary(hours, rate) + bonus


m = Manager()
print("Salary:", m.calculate_salary(40, 50, 5000))
````

````python
#Q15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that calculates the total price of the product.
class Product:
    def __init__(self, name, price, qty):
        self.name = name
        self.price = price
        self.qty = qty

    def total_price(self):
        return self.price * self.qty


p = Product("Laptop", 50000, 2)
print("Total Price:", p.total_price())
````

````python
#Q16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa")


Cow().sound()
Sheep().sound()
````

````python
#Q17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def get_book_info(self):
        return f"{self.title} by {self.author}, published in {self.year}"


b = Book("Python Guide", "John", 2021)
print(b.get_book_info())
````

````python
#Q18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, rooms):
        super().__init__(address, price)
        self.rooms = rooms


m = Mansion("Delhi", 50000000, 12)
print(m.address, m.price, m.rooms)
````
